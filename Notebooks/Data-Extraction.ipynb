## Load

```python
import requests
import pandas as pd

from bs4 import BeautifulSoup
```

## Showroom

```python
url = 'https://www2.hm.com/en_us/men/products/jeans.html'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}

r = requests.get( url, headers=headers )
```

```python
r
```

```python
soup = BeautifulSoup( r.text, 'html.parser' )
```

```python
products = soup.find('ul', class_='products-listing small')
```

```python
products.find_all( 'li', class_='product-item' )
```

```python
products.find_all('h3', class_='item-heading')
```

```python
products.find_all('a', class_='link')[0].get('href')
```

```python
list_products = []
hrefs = []
for i in range( len(products.find_all('a', class_='link')) ):
    product = products.find_all('a', class_='link')[i].get_text()
    href = products.find_all('a', class_='link')[i].get( 'href' )
    link = 'https://www2.hm.com' + href
    list_products.append( product )
    hrefs.append( link )

pd.DataFrame( {'List of Products':list_products,
               'Link of Products':hrefs})
```

## One Product

```python
url = 'https://www2.hm.com/en_us/productpage.0985197001.html'
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}

r = requests.get( url, headers=headers )
```

```python
soup = BeautifulSoup( r.text, 'html.parser' )
```

```python
products_sku = []
links = []

sku = soup.find( 'a', class_='filter-option miniature active' ).get( 'data-articlecode' )
link = 'https://www2.hm.com/en_us/productpage.' + sku +'.html'
products_sku.append( sku )
links.append( link )

all_products_sku = soup.find_all( 'a', class_='filter-option miniature' )

for i in range( len( all_products_sku ) ):
    sku = all_products_sku[i].get( 'data-articlecode' )
    products_sku.append( sku )
    link = 'https://www2.hm.com/en_us/productpage.' + sku +'.html'
    links.append( link )

    
df = pd.DataFrame( {'Products SKU':products_sku,
                    'Links':links} )

df['Links'][0]
```
